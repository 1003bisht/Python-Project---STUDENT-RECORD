```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of occurrences of each ethnic group
group_counts = [len(df[df["EthnicGroup"] == group]) for group in ["group A", "group B", "group C", "group D", "group E"]]

# Create a pie chart
plt.pie(group_counts, labels=["Group A", "Group B", "Group C", "Group D", "Group E"], autopct = "%1.2f%%")
plt.title("Group Distribution")
plt.show()

```
